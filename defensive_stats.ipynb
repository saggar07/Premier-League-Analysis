```python
import pandas as pd
import ast
import numpy as np
import seaborn as sns
from datetime import datetime
import matplotlib.pyplot as plt
from datasets import load_dataset
data = pd.read_csv(r'c:\Users\Uday\Downloads\dataset - 2020-09-24 (1).csv')

data.fillna(0, inplace=True)
data['Age'] = data['Age'].astype('int64')
data['Interceptions'] = data['Interceptions'].astype('int64')
data['Clean sheets'] = data['Clean sheets'].astype('int64')
data['Duels won'] = data['Duels won'].astype('int64')
data
```

```python
data_def = data[data['Position'] == 'Defender']
```

```python
top_interceptors = data_def.loc[data_def.groupby('Name')['Interceptions'].idxmax()]
top_interceptors =  top_interceptors.sort_values(by ='Interceptions', ascending  = False)
top_interceptors = top_interceptors[['Name', 'Interceptions', 'Club']].head(5)
top_interceptors.reset_index(drop=True, inplace = True)
top_interceptors
```

```python
top_clean_sheets = data_def.loc[data_def.groupby('Name')['Clean sheets'].idxmax()]
top_clean_sheets =  top_clean_sheets.sort_values(by ='Clean sheets', ascending  = False)
top_clean_sheets = top_clean_sheets[['Name', 'Clean sheets', 'Club']].head(5)
top_clean_sheets.reset_index(drop=True, inplace = True)
top_clean_sheets
```

```python
top_tacklers = data_def.loc[data_def.groupby('Name')['Tackle success %'].idxmax()]
top_tacklers =  top_tacklers.sort_values(by ='Tackle success %', ascending  = False)
top_tacklers = top_tacklers[['Name', 'Tackle success %', 'Club']].head(5)
top_tacklers.reset_index(drop=True, inplace = True)
top_tacklers
```

```python
top_duels = data_def.loc[data_def.groupby('Name')['Duels won'].idxmax()]
top_duels =  top_duels.sort_values(by ='Duels won', ascending  = False)
top_duels = top_duels[['Name', 'Duels won', 'Club']].head(5)
top_duels.reset_index(drop=True, inplace = True)
top_duels
```

```python

plt.style.use("dark_background")

fig, ax = plt.subplots(3, 1, figsize=(10, 8))
fig.patch.set_facecolor("#0e1117")


# GLOW FUNCTION 
def glow_barh(data, x, y, ax, color):
    # Glow layers
    for alpha, lw in zip([0.15, 0.08, 0.04], [8, 5, 2]):
        ax.barh(
            data[x],
            data[y],
            color=color,
            alpha=alpha,
            edgecolor=color,
            linewidth=lw
        )
    # Main bar
    ax.barh(
        data[x],
        data[y],
        color=color,
        alpha=1
    )

# PLOTS
glow_barh(top_interceptors, "Name", "Interceptions", ax[0], "#2ECC71")
glow_barh(top_clean_sheets, "Name", "Clean sheets", ax[1], "#3498DB")
glow_barh(top_duels, "Name", "Duels won", ax[2], "#F1C40F")

# TITLES 
ax[0].set_title("Interceptions", fontsize=13, weight="bold", pad=10)
ax[1].set_title("Clean Sheets", fontsize=13, weight="bold", pad=10)
ax[2].set_title("Duels Won", fontsize=13, weight="bold", pad=10)

# CLEANUP 
for a in ax:
    a.set_ylabel("")
    a.set_xlabel("")
    a.grid(axis="x", linestyle=":", alpha=0.9)
    a.spines["top"].set_visible(False)
    a.spines["right"].set_visible(False)
    a.spines["left"].set_visible(False)

# MAIN TITLE 
fig.suptitle(
    "Defenders Ranking Based on Key Defensive Attributes",
    fontsize=16,
    weight="bold",
    color="white"
)

fig.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
